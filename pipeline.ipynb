```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, recall_score

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
train=pd.read_csv('Loan_train.csv')
test=pd.read_csv('Loan_test.csv')
sample=pd.read_csv('Loan Sample submission.csv')
```

```python
train.head()
```

```python
train.isna().sum()
```

```python
train['Loan_ID'].nunique()
```

```python
X=train.drop(columns=['Loan_ID','Loan_Status'])
```

```python
X
```

```python
y=train['Loan_Status']
```

```python
X_train , X_val, y_train, y_val = train_test_split(X,y, test_size=0.3, random_state=2)
```

```python
X.columns
```

```python
categorical_features = ['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'Credit_History', 
                        'Property_Area']
numerical_features = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term']
```

```python
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy="most_frequent")),
    ('Onehot', OneHotEncoder())
])
```

```python
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy="median")),
    ('scale', StandardScaler())
])
```

```python
preprocessor = ColumnTransformer(transformers=[('cat_transform',categorical_transformer,categorical_features),
                                 ('num_transform',numerical_transformer,numerical_features),])
```

```python
model_preprocess_pipeline=Pipeline(steps=[('preprocessor',preprocessor),('logistic_model',LogisticRegression())])
```

```python
model_preprocess_pipeline
```

```python
model_preprocess_pipeline.fit(X_train,y_train)
```

```python
model_preprocess_pipeline.predict(X_train)
```

```python
model_preprocess_pipeline.predict_proba(X_train)
```

```python
test_pred=model_preprocess_pipeline.predict(X_val)
```

```python
train_pred = model_preprocess_pipeline.predict(X_train)
```
